First, we note that American Options are at least equally valuable as European Options since the early exercise feature gives flexibility to the holder. 
Thus: 

$C_{A,t} \ge C_{E,t} \tag{1}$ and  $P_{A,t} \ge P_{E,t} \tag{2} $ 

By comparing the payoff of a Long European Call Option with a Long Stock and borrowing PV(K) we get:

|                 | $t$           | $T:S\ge K$ | $T:S<K$ |
|:-----------------:|:----------------:|:---------:|:---------:|
| long Call (1)   | $-C_{E,t}$         | $S_T-K$ | $0$     |
| Long Stock (2)  | $-S_t$         | $S_T$   | $S_T$   |
| Short PV(K) (3) | $PV(K)$        | $-K$    | $-K$    |
|                 |                |         |         |
| 1               |  $-C_{E,t}$        | $S_T-K$ | $0$     |
| 2+3             |  $-S_t + PV(K)$| $S_T-K$ | $S_T-K$ |


By comparing the Pay-Offs at expiry T, we can see immediately that the European Call must be worth equal to or more than the commbination of Buying the Stock and borrowing PV(K) since its payoff is never smaller as it aviods the negative payoff if $S<K$ as the Option simply exipires worthless.

Thus: 
$C_{E,t} \ge S_t-PV(K) \tag{3}$

From (1), we thus receive: 
$C_{A,t} \ge S_t-PV(K) \tag{4}$

By adding an extra value to the option price (denoted as IV = Insurance Value, IV $\ge 0$) we get: 

$C_{A,t} = S_t-PV(K)+IV \tag{5}$

We expand the right hand expression with $\pm$ K and get: 

$C_{A,t} = \underbrace{(S_t-K)}_{Inner Value}+ \underbrace{\big(K-PV(K)\big)}_{Time Value}+\underbrace{IV}_{Insurance Value} \tag{6}$

If we early exercise the American Option, we always receive $S_t-K$, thus losing out on the Time Value and Insurance Value $K-PV(K) + IV\ge 0$. Thus, it is never optimal to early exercise an American Option on an underlying without dividend Payments. (assuming $r\ge 0$)

Thus, for an Option on such an underlying we have: 

$C_{A,t} = C_{E,t}$

Similarly, for a European Put option, we can compare the Payoff of a long Put Option with A short Position in the Underlying and a long Position in PV(K). 

|                 | $t$           | $T:S \ge K$ | $T:S<K$ |
|-----------------|---------------|---------|---------|
| long Put(1)     | $-P_{E,t}$        | $0$     | $K-S_T$ |
| Short Stock (2) | $S_t$         | $-S_T$  | $-S_T$  |
| Long PV(K) (3)  | $-PV(K)$      | $K$     | $K$     |
|                 |               |         |         |
| 1               | $-P_{E,t}$        | $0$     | $K-S_T$ |
| 2+3             | $S_t - PV(K)$ | $K-S_T$ | $K-S_T$ |


Here as well, we can see that the European Put Option is worth at least as much as the combination of Short Stock and Long PV(K) since it avoids the negative payoff for S>K.

Thus: 
$P_{E,t} \ge PV(K) - S_t  \tag{7}$

From (2), we thus receive: 
$P_{A,t} \ge PV(K) - S_t \tag{8}$

Similarly to above, we can modify this to express the value of $P_{A,t}$ as: 

$P_{A,t} = \underbrace{(K-S)}_{Inner Value}+ \underbrace{\big(PV(K)-K\big)}_{Time Value}+\underbrace{IV}_{Insurance Value} \tag{9}$

If we early exercise the Ameican Option on an underlying that does not pay dividend, we receive the inner value and forgo the Time Value and Insurance Value. 

However, we can see that $PV(K) - K + IV$ can also be negative, especially if interest rates are high (and thus PV(K) low) and volatility is also low (and thus IV low). 

Thus, it could be that it pays off to early exercise an American Put Option and therefore, unlike with the Call Option, it does not hold that an American Put on an underlying without dividends necessarily has the same value as its european coutnerparty but that it could potentially be more valuable as it sometimes pay to exercise early

Thus: 

$P_{A,t} \ge  P_{E,t}$

1. $C_{A,t} = C_{E,t}$
2. $P_{A,t} \ge  P_{E,t} \implies P_{E,t} = P_{A,t}  - \epsilon$   , $\epsilon \ge 0$


From the put call parity for european Options (not proven here, as given above) we know that:

3. $ C_{E,t} + \text{PV}(K) = P_{E,t} + S_{t} $

$C_{A,t} + PV(K) = P_{A,t} - \epsilon + S_t \implies C_{A,t} + PV(K) \le P_{A,t}+ S_t$

## Question 5: zipper payoff [15 points]

In this section, you have to first define the payoff function for a call option. By combining the payoffs of call options at different strike prices ranging from 1 to 9, plot the payoff diagram for a "zipper" strategy.

We define the Payoff of a Call Option at expiry as: 

$\text{Payoff}_{Long} = max(S_T-K,0)$

$\text{Payoff}_{Short} = -\text{Payoff}_{Long}$

Thus, we only define the intrinsic value of the Option as Payoff. The Premium paid/ received is not considered as part of the Payoff at Expiry here.
Of course, the total profit at exiry includes the paid/ received premium and is shown in form of the dashed lines below. 

```python
#Import needed libraries
import numpy as np
import matplotlib.pyplot as plt

##################################################################################

#we define an Option Class, which contains all needed Option Parameters
#(Flag (C for Call, P for Put, Strike and the Options Price))
class Option:
    def __init__(self, Flag, Strike, Price = 0):
        self.Flag = Flag
        self.Strike = Strike
        self.Price = Price

    # Calculate the Payoff of an Option over a given Price Range for the Underlying
    # Value = Range over which the Payoff shall be calculated: Range = [Start, End]
    def Payoff(self, Value, Size):

        Value = np.arange(Value[0], Value[1] + 1, 1)
        # Payoff for Calls
        if self.Flag == "C":
            Payoff = np.maximum((Value - self.Strike),0) * Size

        # Payoff for Puts
        elif self.Flag == "P":
            Payoff = np.maximum((self.Strike - Value),0) * Size
            
        return Payoff

    # Calculate the Profit of an Option over a given Price Range for the Underlying
    def Profit(self, Value, Size):

        # Profit for Calls
        if self.Flag == "C":
            Profit = self.Payoff(Value, Size) - (self.Price * Size)

        # Profit for Puts
        elif self.Flag == "P":
            Profit = self.Payoff(Value, Size) - (self.Price * Size)

        return Profit

##################################################################################

#Define necessary Option Parameters, Strike and Price are set randomly 
Call = Option("C",Strike = 50, Price = 8)
Range = [0,100]

#Plot Graphs
fig = plt.figure()
ax = fig.add_axes([0,0,1.6,1.5])
plt.plot(Call.Payoff(Range, Size = 1), color = "r")
plt.plot(Call.Payoff(Range, Size = -1), color = "b")
plt.plot(Call.Profit(Range, Size = 1), "--", color = "r")
plt.plot(Call.Profit(Range,Size =  -1), "--", color = "b")

#Define legend and set titles and axis tick sizes 
plt.legend(["Payoff_Long", "Payoff_Short", "Profit_Long", "Profit_Short"])
ax.set_title("Call Option Long/Short Payoff & Profits", fontsize = 18)
ax.set_xlabel('Stock Price', fontsize=14)
ax.set_ylabel('Profit/Payoff', fontsize=14)

#show plot
plt.show()
```
